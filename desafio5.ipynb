```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd
import numpy as np
import seaborn as sns

# Definindo uma configuração personalizada para a exibição de dados
pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

```python
url = "https://raw.githubusercontent.com/dsreliete/Womakerscode-DA-desafio_squad/refs/heads/main/desafio_01_preco_imoveis.csv"
df = pd.read_csv(url)
df
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y%m%dT%H%M%S')
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
# 2 - Para as top 5 variáveis mais correlacionadas com o preço:

# plote as suas distribuições (histograma e boxplot)

top_5_features = corr["price"].sort_values(ascending=False).index[1:6]

# Criar uma figura com subplots
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(12, 20))

for i, feature in enumerate(top_5_features):
    # Histograma
    sns.histplot(df[feature], ax=axes[i, 0], kde=True)
    axes[i, 0].set_title(f'Histograma de {feature}')

    # Boxplot
    sns.boxplot(x=df[feature], ax=axes[i, 1])
    axes[i, 1].set_title(f'Boxplot de {feature}')

plt.tight_layout()
plt.show()


```
