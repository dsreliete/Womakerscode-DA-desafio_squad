# **Desafio 2: Conversão**

Definição do problema: Dada a tabela com a média de altura e
peso de mulheres estadunidenses, onde os nomes das colunas
estão em inglês height (altura) e weight (peso) e seus
respectivos valores estão nas seguintes unidades: polegada e
libras.
1. Construa uma nova representação onde os nomes das colunas
estão em português, e os valores de altura são convertidos para
centímetros e os valores de peso convertidos para quilograma.
2. Plote o scatterplot de altura x peso da nova representação.

Código do dataset: women

Dicas:

1 polegada = 2,54 centímetros

1 libra = 0,45 quilos

use o método .plot.scatter()

```python
df_mulheres = data('women')
df_mulheres.head()
```

```python
# 1. Construa uma nova representação onde os nomes das colunas estão em português, e os valores de altura são convertidos para
# centímetros e os valores de peso convertidos para quilograma.
df_mulheres_pt = df_mulheres.copy()
```

```python
df_mulheres_pt.rename(columns={'height': 'Altura (cm)', 'weight': 'Peso (kg)'}, inplace=True)
df_mulheres_pt['Altura (cm)'] = df_mulheres_pt['Altura (cm)'] * 2.54
df_mulheres_pt['Peso (kg)'] = df_mulheres_pt['Peso (kg)'] * 0.45
df_mulheres_pt.head()
```

```python
# 2. Plote o scatterplot de altura x peso da nova representação.
from matplotlib import pyplot as plt

df_mulheres_pt.plot.scatter(x='Altura (cm)', y = 'Peso (kg)', title='Altura vs Peso de Mulheres Estadunidenses')
```

# **Desafio 3: Titanic**
Definição do problema:Dado que você está trabalhando com um grupo
 de historiadores que estão analisando o naufrágio do Titanic, informe
 aos historiadores as seguintes questões?

1. Quantas pessoas sobreviveram e quantas morreram?
2. Quantas pessoas eram da primeira, segunda e terceira classe? 3. Qual a porcentagem de sobreviventes na primeira classe?
4. Qual a porcentagem de mortos da terceira classe?


```python
df_titanic = data('titanic', show_doc=True)
df_titanic = data('titanic')

# Verificar a tabela
df_titanic
```

```python
# 1. Contar o número de sobreviventes e mortos

sobreviveram_contagem = df_titanic['survived'].value_counts()

print(sobreviveram_contagem)
```

```python
# 2. Quantas pessoas eram da primeira, segunda e terceira classe?

contagem_de_classe = df_titanic['class'].value_counts()

print(contagem_de_classe)
```

```python
# 3. Qual a porcentagem de sobreviventes na primeira classe?

# Filtrar as pessoas que estavam na primeira classe

df_primeira_classe = df_titanic[df_titanic['class'] == '1st class']

# Contar o número de sobreviventes na primeira classe

sobreviventes_primeira_classe = df_primeira_classe[df_primeira_classe['survived'] == 'yes']

# Calcular a porcentagem de sobreviventes na primeira classe
# O len() é uma função integrada em Python que retorna o número de itens em um objeto

porcentagem_sobreviventes = (len(sobreviventes_primeira_classe) / len(df_primeira_classe)) * 100

# Exibir o resultado

print(f"A porcentagem de sobreviventes na primeira classe é: {porcentagem_sobreviventes:.2f}%")
```

```python
# 4. Qual a porcentagem de mortos da terceira classe?

# Filtrar as pessoas que estavam na terceira classe

df_terceira_classe = df_titanic[df_titanic['class'] == '3rd class']

# Contar o número de mortos da terceira classe

mortos_terceira_classe = df_terceira_classe[df_terceira_classe['survived'] == 'no']

porcentagem_mortos_terceira_classe = (len(mortos_terceira_classe) / len(df_terceira_classe)) * 100

print(f"A porcentagem de mortos na terceira classe é: {porcentagem_mortos_terceira_classe:.2f}%")
```

# **Desafio 4: Futebol**
Definição do problema: Dado que você está trabalhando numa
 emissora de televisão, o narrador precisa informar o jogo com o maior
 número de gols na história da Bundesliga (Campeonato Alemão).
 Você precisa fazer uma busca em todos os jogos da história deste
 campeonato e verificar qual jogo é esse e passar a iformação ao
 narrador, para posteriormente, o narrador informar aos
 telespectadores.
1. Verificar os jogos com maior quantidade de gols na história da
Bundesliga
2. Informar a data, os times participantes e o placar do jogo

```python
import pandas as pd
# Carregar dataset "Bundesliga" e exibir primeiras 10 linhas
df_bundesliga = data('Bundesliga')
```

```python
df_bundesliga.head(10)
```

```python
# 1. Verificar os jogos com maior quantidade de gols na história da Bundesliga

# Criar a coluna TotalGoals
df_bundesliga['TotalGoals'] = df_bundesliga['HomeGoals'] + df_bundesliga['AwayGoals']

#  Encontrar o maior número de gols em um jogo
max_goals = df_bundesliga['TotalGoals'].max()

#  Filtrar os jogos com o maior número de gols e fazer uma cópia do DataFrame
mais_gols = df_bundesliga.loc[df_bundesliga['TotalGoals'] == max_goals].copy()
mais_gols
```

```python
# Informar a data, os times participantes e o placar do jogo

mais_gols['Placar'] = mais_gols['HomeGoals'].astype(str) + ' X ' + mais_gols['AwayGoals'].astype(str)

# Remover a parte da hora na coluna de data
mais_gols['Date'] = pd.to_datetime(mais_gols['Date']).dt.date


#  Formatando o Resultado com alinhamento
mais_gols['Resultado'] = mais_gols.apply(
    lambda row: f"Data: {row['Date']}, jogo {row['HomeTeam'].ljust(30)} {row['Placar'].center(10)} {row['AwayTeam'].rjust(30)}",
    axis=1
)

#  Criar a nova tabela contendo apenas a coluna Resultado
nova_tabela = mais_gols[['Resultado']]


#  Exibir a tabela formatada, removendo o índice
print("JOGOS".center(117))
print(nova_tabela.to_string(index=False, header=False))
```

# **Desafio 5: Forbes**

Definição do problema: Você está trabalhando em um jornal e o editor quer que você responda as seguintes questões:

1. Qual a empresa mais valiosa da lista da Forbes 2000?
2. Construa uma tabela com as TOP 10 empresas mais lucrativas
3. Qual a média de valores de empresa das cinco categorias mais valiosas?

```python
df_forbes = data('Forbes2000')
df_forbes.head()
```

```python
# 1 Qual a empresa mais valiosa da lista da Forbes 2000?
maior_valor_mercado = df_forbes['marketvalue'].max()
df_forbes[df_forbes['marketvalue'] == maior_valor_mercado][['name', 'marketvalue']]
```

```python
# 2 Construa uma tabela com as TOP 10 empresas mais lucrativas da lista.
top_10_empresas_mais_valiosas = df_forbes.nlargest(10, 'profits')[['name', 'profits']]
top_10_empresas_mais_valiosas
```

```python
# 3 Qual a média de valores de empresa das cinco categorias mais valiosas?
media_categoria = df_forbes.groupby('category')['marketvalue'].mean()
media_categoria.sort_values(ascending=False)[0:5]
```
