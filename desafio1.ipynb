```python
df_women = data('women')
```

```python
df_women
```

```python
# Construa uma nova representação onde os nomes das colunas estão em português, e os valores de altura são convertidos para
# centímetros e os valores de peso convertidos para quilograma.

def convert_altura_cm(coluna):
  return coluna * 2.54

def convert_peso_kg(coluna):
  return coluna * 0.45

altura = df_women['height'].agg(convert_altura_cm)
peso = df_women['weight'].agg(convert_peso_kg)

df_women['altura_cm'] = altura
df_women['peso_kg'] = peso

df_women_new = df_women.drop(columns=['height', 'weight'])
df_women_new
```

```python
# Plote o scatterplot de altura x peso da nova representação.

df_women_new.plot.scatter('altura_cm', 'peso_kg')
```

```python
df_titanic = data('titanic')
df_titanic
```

```python
df_titanic.info()
```

```python
# 1. Quantas pessoas sobreviveram e quantas morreram?

result = df_titanic['survived'].value_counts()

print('O número de sobreviventes é ' + str(result[0]))
print('O número de mortos é ' + str(result[1]))
```

```python
# 2. Quantas pessoas eram da primeira, segunda e terceira classe?
classes = df_titanic['class'].value_counts()
classes

first = classes[1]
second = classes[2]
third = classes[0]

print('O número de passageiros da terceira classe é ' + str(third))
print('O número de passageiros da segunda classe é ' + str(second))
print('O número de passageiros da primeira classe é ' + str(first))
```

```python
# 3. Qual a porcentagem de sobreviventes na primeira classe?

first_surv = df_titanic[(df_titanic['class'] == '1st class') & (df_titanic['survived'] == 'yes')].count()[0]
porcentagem1 = (first_surv / first) * 100

print('A porcentagem de sobreviventes na primeira classe é ' + str(porcentagem1) + '%')
```

```python
# 4. Qual a porcentagem de mortos da terceira classe?
third_no_surv = df_titanic[(df_titanic['class'] == '3rd class') & (df_titanic['survived'] == 'no')].count()[0]
porcentagem3 = (third_no_surv / third) * 100

print('A porcentagem de sobreviventes na terceira classe é ' + str(porcentagem3) + '%')
```

```python
df_forbes = data('Forbes2000')
df_forbes
```

```python
# 1. Qual a empresa mais valiosa da lista da Forbes 2000?

df_forbes.sort_values('marketvalue', ascending=False).head(1)
```

```python
# 2. Construa uma tabela com as TOP 10 empresas mais lucrativas da lista.

top10 = df_forbes.nlargest(10, 'profits')
top10['name'].unique()
```

```python
# 3. Qual a média de valores de empresa das cinco categorias mais valiosas?

df_5 = df_forbes.groupby('category')['marketvalue'].mean()
df_5.sort_values(ascending=False).head(5)
```
